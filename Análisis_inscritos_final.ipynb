### BD inscritos 2000 - 2013

Particularmente en la base de inscritos entre los años 2000 y 2013 se evidencia que los inscritos está en columnas, por lo que se realiza un tratamiento especial para hacer unpivot a la base de datos

En el dataframe valores se seleccionan las columnas que permanecerán estáticas y la función pd.melt permite hacer unpivot de la base de datos 

```python
valores = bd2013[bd2013.columns[21:]]
```

```python
base_despiv = pd.melt(bd2013, id_vars = bd2013[bd2013.columns[:21]], value_vars=valores,value_name="INSCRITOS")
```

La nueva variable que contiene la inforamciòn de genero, año y semestre debe ser separada, se conservan las variables que se han separado 

```python
var = base_despiv["variable"].str.split(expand=True)
var.columns = ['GENERO', 'PERIODO']
base_despiv = pd.concat([base_despiv, var], axis=1)
```

```python
var1 = base_despiv["PERIODO"].str.split('-',expand=True)
var1.columns = ['ANIO', 'SEMESTRE']
base_despiv = pd.concat([base_despiv, var1], axis=1)
```

```python
bd_inscritos_2020_2013 = base_despiv.drop(['PERIODO','variable'], axis=1)
bd_inscritos_2020_2013.head()
```

Estandarizar el formato de los datos (dejan en letras capitales, mayúsculas corregir ortografía etc.) y Reducir redundancia en los datos 

```python
for column in inscritos_consol.columns:
    # En mayúsculas columnas categóricas
    if column in cols_cat:
        inscritos_consol[column] = inscritos_consol[column].str.upper()
```

```python
for column in inscritos_consol.columns:
    # En mayúsculas columnas categóricas
    if column in cols_cat:
        inscritos_consol[column] = inscritos_consol[column].str.replace('Á','A', regex=False)
        inscritos_consol[column] = inscritos_consol[column].str.replace('É','E', regex=False)
        inscritos_consol[column] = inscritos_consol[column].str.replace('Í','I', regex=False)
        inscritos_consol[column] = inscritos_consol[column].str.replace('Ó','O', regex=False)
        inscritos_consol[column] = inscritos_consol[column].str.replace('Ú','U', regex=False)
```
